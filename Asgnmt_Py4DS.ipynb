```python
plt.plot(d['PL'])
plt.plot(d['SW'])
plt.plot(d['Classification'])
plt.title('Sebal width VS Petal Lenght')
plt.show()
```

```python
plt.plot(d['Classification'])
plt.title('Petal width VS Petal Lenght')
plt.plot(d['PL'])
plt.plot(d['PW'])
plt.show()
```

```python
clr=['red','green','blue']
Classification=['Iris-setosa','Iris-versicolor','Iris-virginica']
```

```python
for i in range(3):
  x=d[d['Classification']==Classification[i]]
  plt.scatter(x['SL'],x['SW'],c=clr[i],label=Classification[i],marker='^')
  plt.title('Sebal width VS Sebal Lenght')
  plt.xlabel('Sebal Lenght')
  plt.ylabel('Sebal Width')
  plt.legend()
```

```python
for i in range(3):
  x=d[d['Classification']==Classification[i]]
  plt.scatter(x['SW'],x['PW'],c=clr[i],label=Classification[i],marker='^')
  plt.title('Sebal width VS petal width')
  plt.xlabel('Sebal Width')
  plt.ylabel('Petal Width')
  plt.legend()
```
